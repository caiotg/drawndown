```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdate
import matplotlib.ticker as mtick
import mplcyberpunk
import datetime as dt

plt.style.use('cyberpunk')
```

```python
dataInicio = dt.datetime.now() - dt.timedelta(2500)
dataFinal = dt.datetime.now()

ticker = 'TASA4.SA'

cotacoes = yf.download(ticker, dataInicio, dataFinal)['Adj Close']
```

```python
cotacaoMaxima = cotacoes.cummax()
drawndowns = (cotacoes - cotacaoMaxima)/cotacaoMaxima
maxDD = drawndowns.min()

maxDD
```

```python
fig, ax = plt.subplots()
ax.plot(drawndowns.index, drawndowns)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))
ax.xaxis.set_major_locator(mdate.YearLocator(1))
```
